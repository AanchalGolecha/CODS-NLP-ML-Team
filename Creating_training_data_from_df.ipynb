```python
import pickle
import pandas as pd
```

```python
filename='D:\\Python\\Chatbot(rasa)\\Web_scrapping\\Q and A\\Q&A on Be Active during COVID-19.pkl'
f=open(filename,'rb')
df=pickle.load(f)
f.close()

topic='on_physical_health'
df
```

```python
def question_to_intent(df,topic): #nlu.md
    i=1
    for j in df['Questions']:
        print('## intent:'+topic+str(i)+'\n- '+j,end='\n\n')
        i+=1
question_to_intent(df,topic)
```

```python
def intent_to_path(df,topic): #stories.md
    for i in range(1,len(df['Questions'])+1):
        print('## '+topic+str(i)+'_path\n* '+topic+str(i)+'\n - utter_'+topic+str(i),end='\n\n')

intent_to_path(df,topic)
```

```python
def intents(df,topic): #domains.yml
    for i in range(1,len(df['Questions'])+1):
        print('  - '+topic+str(i))
intents(df,topic)
```

```python
def responses(df,topic): #domain.yml
    j=1
    for i in df['Answers']:
        print(' utter_'+topic+str(j)+':\n  - text: "'+i+'"',end='\n\n')
        j+=1
responses(df,topic)
```

```python
def actions(df,topic): #domains.yml
    for i in range(1,len(df['Answers'])+1):
        print('- utter_'+topic+str(i))
actions(df,topic)
```
